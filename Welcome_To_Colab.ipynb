```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def generate_pva_data(num_breaths=5, asynchrony_type=None):
    fs = 100  # Sampling frequency (100Hz)
    t = np.linspace(0, num_breaths * 4, num_breaths * 4 * fs)
    flow = np.zeros_like(t)
    pmus = np.zeros_like(t)

    for i in range(num_breaths):
        start = i * 4 * fs
        ins_end = start + 1 * fs  # 1 second inspiration
        exp_end = (i + 1) * 4 * fs

        # Standard Ventilator Flow (Square Wave)
        flow[start:ins_end] = 0.6
        flow[ins_end:exp_end] = -0.3 * np.exp(-(t[ins_end:exp_end] - t[ins_end]))

        # Inject Asynchrony: Ineffective Trigger during Expiration
        if asynchrony_type == "ineffective_trigger" and i == 2:
            trigger_start = ins_end + 1 * fs
            trigger_end = trigger_start + 0.5 * fs
            pmus[int(trigger_start):int(trigger_end)] = 5 # Patient effort

    volume = np.cumsum(flow) / fs
    # Calculate Pressure: P = (R * Flow) + (V / C) + PEEP - Pmus
    pressure = (10 * flow) + (volume / 0.05) + 5 - pmus

    return pd.DataFrame({'Time': t, 'Pressure': pressure, 'Flow': flow, 'Pmus': pmus})

# Generate the data
df = generate_pva_data(num_breaths=5, asynchrony_type="ineffective_trigger")

# Plotting the result
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True)
ax1.plot(df['Time'], df['Pressure'], color='red', label='Airway Pressure')
ax1.set_ylabel('Pressure (cmH2O)')
ax1.legend()
ax2.plot(df['Time'], df['Flow'], color='blue', label='Flow')
ax2.set_ylabel('Flow (L/s)')
ax2.legend()
plt.xlabel('Time (s)')
plt.show()
```

```
# A simple way to 'detect' that dip in your repository
# We look for a negative slope in pressure during the 'expiratory' time
df['Pressure_Diff'] = df['Pressure'].diff()
asynchrony_events = df[(df['Flow'] <= 0) & (df['Pressure_Diff'] < -0.1)]

print(f"Detected {len(asynchrony_events)} potential asynchrony points!")
```

```
# Tell the computer where 'flow' and 'fs' are coming from
flow = df['Flow'].values
pmus = df['Pmus'].values
fs = 100 # This is our sampling rate from earlier

# Now your existing logic will work:
volume = np.zeros_like(flow)
for i in range(1, len(flow)):
    if flow[i] > 0:
        volume[i] = volume[i-1] + flow[i] * (1/fs)
    else:
        volume[i] = max(0, volume[i-1] + flow[i] * (1/fs))

# Update the pressure in the dataframe
df['Pressure'] = (10 * flow) + (volume / 0.05) + 5 - pmus
```

```
# Add red dots where asynchrony was detected
plt.scatter(asynchrony_events['Time'], asynchrony_events['Pressure'],
            color='yellow', label='Asynchrony Detected', zorder=5)
plt.legend()
```

```
def detect_pva(df):
    # Calculate the slope of the pressure
    df['P_slope'] = df['Pressure'].diff()

    # Logic: If Flow is negative (exhaling) but Pressure Slope is negative (effort),
    # it's an Ineffective Trigger.
    df['Is_Asynchrony'] = (df['Flow'] < -0.05) & (df['P_slope'] < -0.05)

    return df[df['Is_Asynchrony'] == True]

# Run it
results = detect_pva(df)
print(f"Total Asynchrony Duration: {len(results)/100} seconds")
```

```
import matplotlib.pyplot as plt

# 1. Run the detection again to make sure we have the latest results
asynchrony_df = detect_pva(df)

# 2. Create the plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10), sharex=True)

# Pressure Plot
ax1.plot(df['Time'], df['Pressure'], color='gray', alpha=0.5, label='Airway Pressure')
ax1.scatter(asynchrony_df['Time'], asynchrony_df['Pressure'], color='red', s=10, label='Ineffective Trigger Detected')
ax1.set_ylabel('Pressure (cmH2O)')
ax1.set_title('PVA Detection Results')
ax1.legend()

# Flow Plot
ax2.plot(df['Time'], df['Flow'], color='blue', label='Flow')
ax2.set_ylabel('Flow (L/s)')
ax2.set_xlabel('Time (s)')

plt.show()
```

```
def inject_double_trigger(df, breath_index=4):
    # Find the start of the specific breath
    fs = 100
    start_idx = breath_index * 4 * fs

    # Force a second flow pulse immediately after the first one ends
    second_pulse_start = start_idx + int(1.2 * fs)
    second_pulse_end = second_pulse_start + int(0.8 * fs)

    df.loc[second_pulse_start:second_pulse_end, 'Flow'] = 0.6
    return df
```

```
# 1. Generate clean data
df_test = generate_pva_data(num_breaths=10)

# 2. Inject the Double Trigger
df_test = inject_double_trigger(df_test, breath_index=6)

# 3. Run your detector on the new 'dirty' data
results = detect_pva(df_test)

print(f"Algorithm found {len(results)} suspicious points in the double-trigger trial.")
```

```
def count_breaths(df):
    # A breath starts when flow goes from negative/zero to positive
    df['Breath_Start'] = (df['Flow'] > 0) & (df['Flow'].shift(1) <= 0)
    total_breaths = df['Breath_Start'].sum()
    return total_breaths

total = count_breaths(df_test)
```

```
# Calculate the AI%
# We assume if asynchrony points exist within a breath cycle, that breath is asynchronous
asynch_breaths = 1 # In our manual injection case
total_breaths = count_breaths(df_test)

ai_index = (asynch_breaths / total_breaths) * 100
print(f"Final Analysis: Asynchrony Index is {ai_index}%")
```
